```python
PGA_zone = area[area['directory'] == 'PGA_zone']
PGA_zone = PGA_zone[PGA_zone['terrain_variable'] != 'dem']
PGA_zone = PGA_zone[PGA_zone['terrain_variable'] != 'EQIL_area']
PGA_zone = PGA_zone[PGA_zone['terrain_variable'] != 'ground_failure_estimate_cropped']

PGA_zone
```

```python
EQIL_distribution_area = area[area['directory'] == 'EQIL_distribution_area']
EQIL_distribution_area = EQIL_distribution_area[EQIL_distribution_area['terrain_variable'] != 'dem']
EQIL_distribution_area = EQIL_distribution_area[EQIL_distribution_area['terrain_variable'] != 'pga_contour_raster_cropped']
EQIL_distribution_area = EQIL_distribution_area[EQIL_distribution_area['terrain_variable'] != 'ground_failure_estimate_cropped']
EQIL_distribution_area
```

```python
binary_maps = area[area['directory'] == 'binary_maps']
binary_maps = binary_maps[binary_maps['terrain_variable'] != 'dem']
binary_maps = binary_maps[binary_maps['terrain_variable'] != 'EQIL_area']
binary_maps = binary_maps[binary_maps['terrain_variable'] != 'ground_failure_estimate_cropped']
binary_maps = binary_maps[binary_maps['terrain_variable'] != 'pga_contour_raster_cropped']

binary_maps
```

We can discard dem, EQIL area, PGA
We are interested in distance_stream_downstream, local_relief, profile_curvature, slope, TWI, VRM

```python
PGA_zone_profile_curvature = PGA_zone[PGA_zone['terrain_variable'] == 'profile_curvature'].mean_value
ground_failure_zone_profile_curvature = ground_failure_zone[ground_failure_zone['terrain_variable'] == 'profile_curvature'].mean_value
EQIL_distribution_area_profile_curvature = EQIL_distribution_area[EQIL_distribution_area['terrain_variable'] == 'profile_curvature'].mean_value
binary_maps_profile_curvature = binary_maps[binary_maps['terrain_variable'] == 'profile_curvature'].mean_value
PGA_zone_profile_curvature, ground_failure_zone_profile_curvature, EQIL_distribution_area_profile_curvature, binary_maps_profile_curvature
```

Profile curvature is a measure describing the concavity/convexity of slope along the vertical direction. A positive value indicates concave slope, while a negative value indicated a convex one. Having a concave surface can increase slope instability by increasing the subsurface drainage that can cause high water pressure (Pierson 1980). By analyzing the profile curvature mean we can see on average the slope stability. Given the results above, we can see tha the PGA_zone, ground_failure_Zone and EQIL_distribution area are concave, although the latter two are extremely small compared to the first one. On the other hand, the binary map, which is the entire area, seems to be slightly convex

```python
PGA_zone_VRM2 = PGA_zone[PGA_zone['terrain_variable'] == 'vector_ruggedness_measure'].min_value
ground_failure_zone_VRM2  = ground_failure_zone[ground_failure_zone['terrain_variable'] == 'vector_ruggedness_measure'].min_value
EQIL_distribution_area_VRM2  = EQIL_distribution_area[EQIL_distribution_area['terrain_variable'] == 'vector_ruggedness_measure'].min_value
binary_maps_VRM2  = binary_maps[binary_maps['terrain_variable'] == 'vector_ruggedness_measure'].min_value

PGA_zone_VRM2 , ground_failure_zone_VRM2 , EQIL_distribution_area_VRM2 , binary_maps_VRM2 
```

Tanyas et al. (2019) showed that the majority of EQIL are initiated at low VRM values. Furthemore, the lower the VRM value is then the chances of a landslide occuring is higher. In this case, if one area has drastically lower VRM than the others then it means that area is most likely to have an earthquake occur. Thus, we are looking for the lowest value possible in all areas/maps. As observed, the binary map is the only one with a negative value, although the others are positive but still very close to 0. The discrepancy is not too considerable, but it shows the ara most likely to have a landslide occur

```python
PGA_zone_local = PGA_zone[PGA_zone['terrain_variable'] == 'local_relief'].max_value
ground_failure_zone_local  = ground_failure_zone[ground_failure_zone['terrain_variable'] == 'local_relief'].max_value
EQIL_distribution_area_local  = EQIL_distribution_area[EQIL_distribution_area['terrain_variable'] == 'local_relief'].max_value
binary_maps_local  = binary_maps[binary_maps['terrain_variable'] == 'local_relief'].max_value

PGA_zone_local , ground_failure_zone_local , EQIL_distribution_area_local , binary_maps_local 
```

Local relied is the maximum difference in height in a local neighborhood of each pixel and can be related to slope instability caused by tectonic uplift. Thus, the maximum value in those maps indicates the maximum difference in height and a higher slope instablity. As it can be seen, all  maps have the same value of 9.0, with exception of ground_failure_zone_local which is quite lower than the others at 3.0 . 

```python
PGA_zone_TWI = PGA_zone[PGA_zone['terrain_variable'] == 'TWI_raster'].standard_deviation_unbiased
ground_failure_zone_TWI  = ground_failure_zone[ground_failure_zone['terrain_variable'] == 'TWI_raster'].standard_deviation_unbiased
EQIL_distribution_area_TWI  = EQIL_distribution_area[EQIL_distribution_area['terrain_variable'] == 'TWI_raster'].standard_deviation_unbiased
binary_maps_TWI  = binary_maps[binary_maps['terrain_variable'] == 'TWI_raster'].standard_deviation_unbiased

PGA_zone_TWI , ground_failure_zone_TWI , EQIL_distribution_area_TWI , binary_maps_TWI
```

TWI (Moore et al. 1991) is a proxy for potential soil wetness used to estimate the spatial variability of wetness within a landscape (e.g., Nowicki Jessee et al. 2018). It can take a role in slope stability by changing the pore water pressure. By looking at the standard deviation of the TWI of our maps, we can see the values are quite high in three of the cases, which means the values are quite scattered around the means in great distances.

```python
PGA_zone_slope = PGA_zone[PGA_zone['terrain_variable'] == 'slope'].median_value
ground_failure_zone_slope  = ground_failure_zone[ground_failure_zone['terrain_variable'] == 'slope'].median_value
EQIL_distribution_area_slope  = EQIL_distribution_area[EQIL_distribution_area['terrain_variable'] == 'slope'].median_value
binary_maps_slope  = binary_maps[binary_maps['terrain_variable'] == 'slope'].median_value

PGA_zone_slope , ground_failure_zone_slope , EQIL_distribution_area_slope , binary_maps_slope
```

Slope is a factor controlling the normal and shear stresses, which take a role in slope stability. Budimir et al. (2015) stated that in all those studies slope was found as a significant variable. There is quite a discrepancy of the median value between the four maps. The PGA_zone and binary_maps seems to have quite similar values between 1 and 2, whereas the EQIL_distribution_area has higher values at around the 14 mark, followed by the ground_failure_zone with the highest value of almost 25. That makes sense for the ground_failure_zone since the slope is a very important parameter and a higher slope indicated higher chances of a landslide.

```python
PGA_zone_distance = PGA_zone[PGA_zone['terrain_variable'] == 'distance_stream_downstream'].mean_value
ground_failure_zone_distance  = ground_failure_zone[ground_failure_zone['terrain_variable'] == 'distance_stream_downstream'].mean_value
EQIL_distribution_area_distance = EQIL_distribution_area[EQIL_distribution_area['terrain_variable'] == 'distance_stream_downstream'].mean_value
binary_maps_distance  = binary_maps[binary_maps['terrain_variable'] == 'distance_stream_downstream'].mean_value

PGA_zone_distance , ground_failure_zone_distance , EQIL_distribution_area_distance , binary_maps_distance
```

Distance to stream is proxy related to fluvial undercutting (e.g., Kritikos et al. 2015) that cause high rates of shear stress as a result of loss of lateral support (Korup 2004). Tanya et al. (2017) showed that the majority of EQIL are initiated close to river channels and the frequency of observed landslides gradually decreases while going far away from channels. The lower the distance_stream_downstream, the lower the distance to rive channels which results in higher chances of occuring landslides.The PGA_zone has the lowest distance_stream_downstream, closely followed by he ground_failure_zone and binary_maps which have very similar. Those have considerably higher chances of occuring landslidse compared to the EQIL_distribution_area as they are closer to river channels.
